```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample to test for normality
T1 = [100, 108, 119, 127, 132, 135, 136, 164]
T1.sort()
quantile_levels1 = np.arange(len(T1),dtype=float)/len(T1)
quantiles1 = T1

#Calculate quantiles by generating samples from hypothetical normal distribution
T2 = np.random.normal(np.mean(T1), np.std(T1), 1000)
T2.sort()
quantile_levels2 = np.arange(len(T2),dtype=float)/len(T2)

#We find the set of quantiles for the larger data set using linear interpolation
quantiles2 = np.interp(quantile_levels1, quantile_levels2, T2)
#Plot the quantiles to create the qq plot
plt.plot(quantiles2, quantiles1, 'o')

#Add a reference line
maxval = max(T1[-1], T2[-1])
minval = min(T1[0], T2[0])
plt.plot([minval,maxval],[minval,maxval],'k--');
```
